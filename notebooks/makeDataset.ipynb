```python
data = pd.read_csv("notebooks/data/bq-results-20240502-134940-1714657944585.csv")
```

```python
df = pd.DataFrame(data)
df.head(10)
```

```python
def split_synonyms(synonyms):
    return synonyms.split(',')
```

```python
df['synonyms'] = df['synonyms'].apply(split_synonyms)
```

```python
def generate_positive_pairs(df):
    positive_pairs = []
    for _ in range(len(df)):
        concept_id = random.choice(df['conceptId'])
        terms = random.sample(df[df['conceptId'] == concept_id]['synonyms'].iloc[0], 2)
        positive_pairs.append([terms[0], terms[1], True])
    return positive_pairs
```

```python
def generate_negative_pairs(df):
    negative_pairs = []
    for _ in range(len(df)):
        concept_ids = random.sample(list(df['conceptId']), 2)
        terms_1 = random.choice(df[df['conceptId'] == concept_ids[0]]['synonyms'].iloc[0])
        terms_2 = random.choice(df[df['conceptId'] == concept_ids[1]]['synonyms'].iloc[0])
        negative_pairs.append([terms_1, terms_2, False])
    return negative_pairs
```

```python
positive_pairs = generate_positive_pairs(df)
negative_pairs = generate_negative_pairs(df)
```

```python
pairs = positive_pairs + negative_pairs
```

```python
random.shuffle(pairs)
```

```python
dataset = pd.DataFrame(pairs, columns=['term1', 'term2', 'Synoniem'])
```
